```python
df = pd.read_csv("sample_data/wine-ratings.csv", index_col=0) # read the csv file and set the index column to 0
df.head(5)
```

```python
# Se eliminaran la filas que tengan valores nulos en region
df = df.dropna(subset=['region'])

# Se eliminaran la filas que tengan valores nulos en notes
df = df.dropna(subset=['notes'])
```

```python
austria_df=df[df['region'].str.contains('Austria')]
```

```python
austria_df.columns
```

## Filtrar por Tipo de Vino

```python
white_wines = df[df['variety'] == 'White Wine']
white_wines.head(5)
```

```python
white_wines.to_json('sample_data/white_wines.json', orient='records')
```

## Filtrar por Puntuación Alta

```python
# Obtener todos los vinos con una puntuación igual o superior a 90.
high_rated_wines = df[df['rating'] >= 90]
high_rated_wines.head(5)
```

```python
high_rated_wines.to_json('sample_data/high_rated_wines.json', orient='records')
```

## Filtrar por Región de Austria

```python
austria_wines = df[df['region'].str.contains('Austria')]
austria_wines.head(5)
```

```python
austria_wines.to_json('sample_data/austria_wines.json', orient='records')
```

## Filtrar por Notas con "fruit bouquet"

```python
fruit_bouquet_wines = df[df['notes'].str.contains('fruit bouquet')]
fruit_bouquet_wines.head(5)
```

```python
fruit_bouquet_wines.to_json('sample_data/fruit_bouquet_wines.json', orient='records')
```
